```
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import inv 
```

$$
Calculating\;optimal\;weight\;matrix\;for\;regression\;problem\;with\;Root\;mean\;square\;error.
\\W^{*}=(X^{T}.X)^{-1}.X^{T}.Y
$$

```
N=100
x_g=np.linspace(0,2*np.pi,N)
y=np.sin(x_g)
mean=0
std=0.05
y+=np.random.normal(mean,std,N)
x=np.vstack((np.ones(N),x_g))
x_p=x
x=np.transpose(x)
w=np.linalg.inv(x_p@x)@x_p@y
```

```
plt.xlabel("Training Data(x)")
plt.ylabel("Training label(y)")
plt.title("Vanilla Regression")
plt.plot(x_g,y,'-bo',label='samples of sinefunction')
plt.plot(x_g,x@w,'m:',label='vanilla linear regression')
plt.legend()
plt.show()
```
