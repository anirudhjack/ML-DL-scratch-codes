\\C_{XX}\;is\;a\;symmetric\;matrix\implies C_{XX}=E.D.E^{T}(from\;SVD)
\\where\;E\;is\;a\;matrix\;containing\;eigen\;vectors\;of\;C_{XX}\;and\;D\;is\;a\;diagonal\;matrix\;with\;eigen\;values\;as\;its\;diagonal\;elements.
\\\implies C_{YY}=\frac{1}{N}Y.Y^{T}
\\\implies C_{YY}=\frac{1}{N}(P.X)(P.X)^{T}
\\\implies C_{YY}=\frac{1}{N}(P.X.X^{T}.P^{T})
\\\implies C_{YY}=(P.\frac{1}{N}(X.X^{T}).P^{T})
\\\implies C_{YY}=P.C_{XX}.P^{T}
\\\implies C_{YY}=P.(E.D.E^{T}).P^{T}
\\\implies for\;decorrelating\;the\;data(X)\;C_{YY}\;should\;be\;a\;diagonal\;matrix.
\\\implies P=E^{T}(since\;E.E^{T}=I)
\\\implies C_{YY}=D\;if\;P=E^{T}.

```
import numpy as np
import matplotlib.pyplot as plt
import math
import random as rd
```

```
def determinant(sigma):
  det=1
  for i in range(np.shape(sigma)[0]):
    det*=sigma[i][i]
  return det
```

```
def multivariate_guassian(x,mu,sigma):
  x,mu,sigma=np.array(x),np.array(mu),np.array(sigma)
  term =(-1/2)*np.transpose(x-mu)@np.linalg.inv(sigma)@(x-mu)
  guassian=1/np.sqrt(((2*np.pi)**(np.shape(x)[0]))
                     *determinant(sigma))*math.exp(term)
  return guassian
```

```
def generate_parameters(dim,K):
  mus=list()
  for i in range(K):
    mus.append(np.random.rand(dim))
    
  sigmas=list()
  for i in range(K):
    sigmas.append(np.diag(np.random.rand(dim)))
  
  pis=np.random.dirichlet(np.ones(K))
  
  return (np.array(mus),np.array(sigmas),np.array(pis))
```

```
#r(z_nk)=>n=data,k=index.
def r(point,mus,sigmas,pis,index):
  numerator=pis[index]*multivariate_guassian(point,mus[index],sigmas[index])
  denominator=0
  for i in range(np.shape(mus)[0]):
    denominator+=pis[i]*multivariate_guassian(point,mus[i],sigmas[i])
  
  return (numerator/denominator)
```

```
def Expectation_maximization(data_matrix,K,threshold):
  (mus,sigmas,pis)=generate_parameters(np.shape(data_matrix)[0],K)
  
  err=math.inf
  while(err>=threshold):
    mus_new=list()

    for j in range(K):
      sum1=0
      mu=np.zeros(np.shape(data_matrix)[0])
      for i in range(np.shape(data_matrix)[1]):
        mu=mu+r(data_matrix[:,i],mus,sigmas,pis,j)*data_matrix[:,i]
        sum1+=r(data_matrix[:,i],mus,sigmas,pis,j)
      mus_new.append(mu/sum1)

    sigmas_new=list()

    for j in range(K):
      sum2=0
      sigma=np.zeros((np.shape(data_matrix)[0],np.shape(data_matrix)[0]))
      for i in range(np.shape(data_matrix)[1]):
        sigma=sigma+r(data_matrix[:,i],mus,sigmas,pis,j)*((data_matrix[:,i]-mus_new[j])@np.transpose(data_matrix[:,i]-mus_new[j]))
        sum2+=r(data_matrix[:,i],mus,sigmas,pis,j)
      sigmas_new.append(np.diag(np.diag(sigma))/sum2)

    pis_new=list()
    for j in range(K):
      sum3=0
      for i in range(np.shape(data_matrix)[1]):
        sum3+=r(data_matrix[:,i],mus,sigmas,pis,j)
      pis_new.append(sum3/np.shape(data_matrix)[1])
    
    err=np.sum(np.array(mus)-np.array(mus_new))**2+np.sum(np.array(sigmas)-np.array(sigmas_new))**2+np.sum(np.array(pis)-np.array(pis_new))**2
    print("loss:",end=" ")
    print(err)
    
    mus=mus_new
    sigmas=sigmas_new
    pis=pis_new
  return (mus,sigmas,pis)
```

```
def generate_data(a):
  data2=list()
  clas2=list()
  
  r=1
  while(np.shape(data2)[0]<50):
    x=rd.uniform(a-1,a+1)
    y=rd.uniform(a-1,a+1)
    v=[x,y]

    if (x-a)**2+(y-a)**2<r**2:
      data2.append(v)
      clas2.append(1)


  while(np.shape(data2)[0]<100):
    x=rd.uniform(-a-1,-a+1)
    y=rd.uniform(-a-1,-a+1)
    v=[x,y]

    if (x+a)**2+(y+a)**2<r**2:
      data2.append(v)
      clas2.append(-1)

  while(np.shape(data2)[0]<150):
    x=rd.uniform(-a-1,a+1)
    y=rd.uniform(-a-1,a+1)
    v=[x,y]

    if (x+a)**2+(y-a)**2<r**2:
      data2.append(v)
      clas2.append(2)
      
  while(np.shape(data2)[0]<200):
    x=rd.uniform(a+1,a-1)
    y=rd.uniform(-a+1,-a-1)
    v=[x,y]
    if (x-a)**2+(y+a)**2<r**2:
      data2.append(v)
      clas2.append(-2)

  return (data2,clas2)
```
