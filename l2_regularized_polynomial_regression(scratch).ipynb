```
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import inv 
```

$$
Calculating\;optimal\;weight\;matrix\;for\;l2\;regularized\;regression\;problem\;with\;Root\;mean\;square\;error.
\\W^{*}=(X^{T}.X+\lambda.I)^{-1}.X^{T}.Y
$$

```
def generate_data(Samples,mean,deviation):
  N=Samples
  x_g=np.linspace(0,2*np.pi,N)
  y=np.sin(x_g)
  m=mean
  std=deviation
  y+=np.random.normal(m,std,N)

  x=np.vstack((np.ones(N),x_g))

  return(x,y,x_g)
```

```
def plot_function(x_g,y,x,w,d):
  plt.xlabel("Training Data(x)")
  plt.ylabel("Training label(y)")
  plt.title("Polynomial regression with dimension %d"%(d))
  plt.plot(x_g,y,'-bo',label='samples of function')
  plt.plot(x_g,x@w,'r:',label='polynomial basis function')
  plt.legend()
  plt.show()
```

```
def Polynomial_regression(dimension,samples,Mean,Deviation,lamda):
  (x,y,x_g)=generate_data(Samples=samples,mean=Mean,deviation=Deviation)
  d=dimension
  for i  in range(2,d):
      x= np.vstack((x,np.power(x_g, i)))
  x_p=x
  x=np.transpose(x)
  w=np.linalg.inv(x_p@x+lamda*np.identity(d))@x_p@y
  plot_function(x_g,y,x,w,d)
```

$$
Ridge\;regression\;with\;dimension\;D=14\;without\;regularization
$$

```
Polynomial_regression(dimension=14
                      ,samples=20
                      ,Mean=0.0
                      ,Deviation=0.05
                      ,lamda=0)
```
